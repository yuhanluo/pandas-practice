```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("round1_ijcai_18_train_20180301.txt", sep=' ', encoding='utf-8-sig')
```

# Question 1

```python
corr_table = df.corr()
corr_table['is_trade'] = corr_table['is_trade'].apply(lambda x: abs(x))
corr_table = corr_table.sort_values(by="is_trade", ascending=False)
corr_table.head()
```

# Question 2

```python
df['context_timestamp'] = df['context_timestamp'].apply(lambda x : pd.to_datetime(x, unit='s').replace(hour=0, minute=0, second=0))
```

```python
df1 = df.sort_values(by="context_timestamp", ascending=False)
df1[['is_trade','context_timestamp']].head()
```

```python
from datetime import timedelta
```

```python
tmp = df[df['context_timestamp'] >= df['context_timestamp']- timedelta(2)].groupby(['item_id','context_timestamp']).count().reset_index()
```

```python
tmp[['is_trade','context_timestamp','item_id']].head()
```

```python
tmp1 = df[df['context_timestamp'] >= df['context_timestamp']- timedelta(2)].groupby(['item_id','context_timestamp']).sum().reset_index()
```

```python
tmp1[['is_trade','context_timestamp','item_id']].head()
```

```python
tmp['historical_conversion_rate'] = tmp1['is_trade'] / tmp['is_trade']
tmp['historical_conversion_rate'].head()
```

```python
tmp.reset_index(level=0, inplace=True)
tmp.head()
```

```python
df = pd.merge(df, tmp[['historical_conversion_rate', 'item_id']], on='item_id', how='left')
```

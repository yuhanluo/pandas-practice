# Question 2

```python
df['context_timestamp'] = df['context_timestamp'].apply(lambda x : pd.to_datetime(x, unit='s'))
```

```python
df1 = df.sort_values(by="context_timestamp", ascending=False)
```

```python
from datetime import datetime, timedelta
date = df1['context_timestamp'].max()
start_date = date - timedelta(2)
```

```python
tmp = df[df['context_timestamp'] >= start_date].groupby('item_id').count()
tmp['is_trade'].head()
```

```python
tmp1 = df[df['context_timestamp'] >= start_date].groupby('item_id').sum()
tmp1['is_trade'].head()
```

```python
tmp['historical_conversion_rate'] = tmp1['is_trade'] / tmp['is_trade']
tmp['historical_conversion_rate'].head()
```

```python
tmp.reset_index(level=0, inplace=True)
tmp.head()
```

```python
df = pd.merge(df, tmp[['historical_conversion_rate', 'item_id']], on='item_id', how='left')
```
